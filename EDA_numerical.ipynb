# 3. Data Cleaning
Rename columns:

```python
df.rename(columns={"class" : "Type"}, inplace=True)
```

```python
df.columns
```

 Remove duplicates:

```python
df=df.drop_duplicates()
```

```python
df["Type"]=df["Type"].astype(int)
```

```python
df["Type"].dtype
```

```python

# Sample data
data = {'Type': ["'1'", "'2'", " 3", "'4'", "five"]}
df = pd.DataFrame(data)

print("Before Cleaning:")
print(df)

# Step 1: Remove single quotes and extra spaces
df['Type'] = df['Type'].str.replace("'", "", regex=False).str.strip()

# Step 2: Convert to numeric, setting invalid values to NaN
df['Type'] = pd.to_numeric(df['Type'], errors='coerce')

# Step 3: Check for NaN (optional)
print("Invalid values converted to NaN:")
print(df[df['Type'].isna()])

# Step 4: Drop or fill NaN, then convert to int
df = df.dropna(subset=['Type'])  # Drop rows with NaN
df['Type'] = df['Type'].astype(int)

print("After Cleaning:")
print(df)
```
